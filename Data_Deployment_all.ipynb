```
import os
len(os.listdir('/content/cleaned_data'))
a = 0
b = 0
def countfile(dir_path):
  global a
  global b
  for i in os.listdir(dir_path):
    full_path = dir_path + "/" + i
    if os.path.isfile(full_path):
      a+=1
    else:
      b+=1
      countfile(full_path)

countfile('/content/cleaned_data')

print('總共: ' + str(b) + ' 個字(資料夾) / 總共: ' + str(a) + '個樣本')
print('平均每個字有: ' + str(a/b) + ' 個樣本')
```
